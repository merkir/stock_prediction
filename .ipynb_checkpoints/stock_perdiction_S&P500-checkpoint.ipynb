# 2.2 LSTM

```python
# Scale data
open = DF.filter(['Open'])
openData = np.array(open)
SCALER = MinMaxScaler(feature_range=(0,1))
scaled_data = SCALER.fit_transform(openData)
```

```python
# Create training and testing data

LOOKBACK_DAYS = 60 #for each stock price, we look back 60 days for its perdiction

# Create datasets
DATA_LEN = 1564
TRAIN_DATA_LEN = math.ceil(DATA_LEN*.8)
```

```python
def createXyData(lookbackdays, data):
  x_data=[] #lookbackdays-day perdiction data
  y_data=[] #actual opening price

  for i in range(lookbackdays, len(data)):
    x_data.append(data[i-lookbackdays:i,0])
    y_data.append(data[i])

  return x_data, y_data

def createDataSet(all_data, param, plt):
  start_index, end_index = int(param['start_index']), int(param['end_index'])
  train_len = math.ceil((end_index - start_index)*.8)
  # train data
  train_data = all_data[start_index:start_index+train_len,:]
  x_train, y_train = createXyData(LOOKBACK_DAYS, train_data)
  x_train, y_train = np.array(x_train), np.array(y_train)
  x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
  print(x_train.shape, y_train.shape)

  # test data
  test_data = scaled_data[start_index+train_len-LOOKBACK_DAYS:start_index+(end_index - start_index),:]
  x_test, y_test = createXyData(LOOKBACK_DAYS, test_data)
  x_test, y_test = np.array(x_test), np.array(y_test)
  x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
  print(x_test.shape, y_test.shape)

  plt.plot(open[start_index:start_index+train_len], color='yellow') # train
  plt.plot(open[start_index+train_len:end_index], color='green') # test

  param['x_train'] = x_train
  param['y_train'] = y_train
  param['x_test'] = x_test
  param['y_test'] = y_test

  return param, plt
```

```python
start_index = 0
end_index = len(open)
param_all = {
    'start_index': start_index,
    'end_index': end_index,
}
param_all, plt = createDataSet(scaled_data, param_all, plt)
```

```python
plt.figure(figsize=(15,7))
plt.title('S&P 500 Opening Price')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Open Price USD($)', fontsize=18)

# DATA SET 1: 2003-2009 => 2008-2009
start_index_1 = 0
end_index_1 = start_index_1 + DATA_LEN
param_1 = {
    'start_index': start_index_1,
    'end_index': end_index_1,
}
param_1, plt = createDataSet(scaled_data, param_1, plt)

# DATA SET 2: 2009-2015 => 2014-2015
start_index_2 = 1690
end_index_2 = start_index_2 + DATA_LEN
param_2 = {
    'start_index': start_index_2,
    'end_index': end_index_2,
}
param_2, plt = createDataSet(scaled_data, param_2, plt)

# # DATA SET 3: 2017-2023 => 2022-2023
start_index_3 = 3600
end_index_3 = start_index_3 + DATA_LEN
param_3 = {
    'start_index': start_index_3,
    'end_index': end_index_3,
}
param_3, plt = createDataSet(scaled_data, param_3, plt)

plt.plot(open[end_index_1:start_index_2], color='black')
plt.plot(open[end_index_2:start_index_3], color='black')
plt.plot(open[end_index_3:], color='black')

plt.legend(['Train', 'Test'], loc='best')
# plt.savefig('/content/dataset.png')
plt.show()
```

### LSTM

```python
def lstmExperiment(param, lstm, batch_size=32,epochs=5):
  start_index, end_index, x_train, y_train, x_test, y_test = param['start_index'], param['end_index'], param['x_train'], param['y_train'], param['x_test'], param['y_test']
  lstm, exec_time = lstmTrain(lstm, x_train, y_train, batch_size, epochs)
  mae, mse, rmse, predictions = lstmTest(lstm, x_test, y_test)

  print("execuation time:", exec_time, "mae:", mae, "mse:", mse, "rmse:", rmse)

  # visualize
  train_len = math.ceil((end_index - start_index)*.8)
  train = open[start_index:start_index+train_len]
  valid = open[start_index+train_len:end_index]
  valid['predictions'] = predictions
  plt_all = lstmVisualize(train, valid)
  return lstm, exec_time, predictions, plt



def lstmTrain(lstm, x_train, y_train, batch_size=32,epochs=5):
  lstm.compile(optimizer='adam', loss='mean_squared_error')
  start = time.time()
  lstm.fit(x_train, y_train, batch_size=batch_size,epochs=epochs)
  exec_time = time.time() - start
  return lstm, exec_time

def lstmTest(lstm, x_test, y_test):
  predictions=lstm.predict(x_test)
  mae=np.mean(np.abs(predictions-y_test))
  mse=np.mean((predictions-y_test)**2)
  rmse=np.sqrt(mse)
  print("STANDARLIZED", "mae:", mae, "mse:", mse, "rmse:", rmse)

  predictions=SCALER.inverse_transform(predictions)
  y_test=SCALER.inverse_transform(y_test)
  mae=np.mean(np.abs(predictions-y_test))
  mse=np.mean((predictions-y_test)**2)
  rmse=np.sqrt(mse)
  return mae, mse,rmse, predictions

def lstmVisualize(train, valid):
  #Visualize the data
  plt.figure(figsize=(15,7))
  plt.xlabel('Date', fontsize=18)
  plt.ylabel('Open Price USD($)', fontsize=18)
  plt.plot(train , color='red')
  plt.plot(valid['Open'] , color='blue')
  plt.plot(valid['predictions'] , color='green')
  plt.legend(['Validation', 'predictions'], loc='lower right')
  return plt
```

```python
shape = param_all['x_train'].shape
lstm = Sequential([
      LSTM(100, input_shape=(shape[1],1)),
      Dense(25, activation='relu'),
      Dense(1)
])
```

```python
# all data
lstm_all, exec_time_all, predictions_all, plt_all = lstmExperiment(param_all, lstm, epochs = 10)
plt_all.title('Perdiction of 2020-2023')
plt_all.savefig('/content/prediction_all.png')
plt_all.show()
```
